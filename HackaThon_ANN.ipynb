```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(categorical_features = [0])
y = onehotencoder.fit_transform(y.reshape(-1, 1)).toarray()
```

```python
'''splitting up the dataset'''
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
'''Scaling the result'''
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
Xtrain = sc_X.fit_transform(Xtrain)
Xtest = sc_X.transform(Xtest)
```

```python
plt.subplots(figsize=(20,15))
sns.heatmap(dataset.corr(), annot = True, linewidths = 0.5)
```

```python
# applying the deep learning layers
import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dropout
```

```python
''' making the object'''
# initialising the sequential
classifier = Sequential()
```

```python
# adding the 1st layer and the hidden layer
classifier.add(Dense(16, input_dim = 17, activation = 'relu', kernel_initializer='glorot_uniform'))
classifier.add(Dropout(rate = 0.1))
```

```python
# adding the first layer
classifier.add(Dense(16, kernel_initializer='glorot_uniform', activation = 'relu'))
classifier.add(Dropout(rate = 0.1))
```

```python
# adding the second layer
classifier.add(Dense(16, kernel_initializer = 'glorot_uniform', activation = 'relu'))
classifier.add(Dropout(rate = 0.1))
```

```python
classifier.add(Dense(16, kernel_initializer = 'glorot_uniform', activation = 'relu'))
classifier.add(Dropout(rate = 0.1))
```

```python
classifier.add(Dense(16, kernel_initializer = 'glorot_uniform', activation = 'relu'))
classifier.add(Dropout(rate = 0.1))
```

```python
# adding the final layer
classifier.add(Dense(4, kernel_initializer = 'glorot_uniform', activation = 'softmax'))
```

```python
# compliling the results
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# fitting the model
classifier.fit(Xtrain, ytrain, batch_size = 12,  epochs = 200)
print('***************************************************\n**********************************************')
```

```python
# predicting the results
y_pred = classifier.predict_classes(Xtest)
```

```python
y_pred_df = pd.DataFrame(y_pred)
y_pred_df.head(10)
```

```python
classifier.test_on_batch(Xtest, ytest)
```
