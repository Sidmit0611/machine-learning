## deviding the dataset into features and output feature

#### we have to devide the dataset into our features and target values 
#### features are those values in dataset because of which our output value comes

```python
X = dataset.iloc[:, :-2].values
y = dataset.iloc[:, -2].values
```

```python
X
```

```python
print(X.shape)
print(dataset.shape)
y
```

```python
print(dataset.shape)
print(X.shape)
print(y.shape)
```

```python
# splitting up the dataset
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(dataset.shape)
print(X.shape)
print(y.shape)
print(Xtrain.shape)
print(Xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
# preprocessing
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
Xtrain = sc_X.fit_transform(Xtrain)
Xtest = sc_X.transform(Xtest)
```

```python
# fitting the model into linear regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(Xtrain, ytrain)
```

```python
lin_intercept = regressor.intercept_
lin_coef = regressor.coef_
print(lin_intercept)
print('\n')
print(lin_coef)
```

```python
# predicting the results
y_pred = regressor.predict(Xtest)
```

```python
print(y_pred.shape)
print(ytest.shape)
```

```python
print(y_pred)
```

```python
print(ytest)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(ytest, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(ytest, y_pred)
```

```python
accuracy = regressor.score(Xtest, ytest)
print(accuracy)
```

```python
accuracy_train = regressor.score(Xtrain, ytrain)
print(accuracy_train)
```
